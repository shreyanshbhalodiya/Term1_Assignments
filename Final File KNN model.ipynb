```python
def aut(file):
    try:
        print('hello i am knn model ')
        import pandas as pd
        import numpy as np
        print('\n')
        print('Please remove all null values in files')
              ## file name with zero null values is expected 

                                              ## dropping 0  values rows from file

        print('\n')
        for i in file.columns:      ## printing column names
            print(i , end = ' , ')

        print('\n')

        print("enter the same name of predictor and target variable to avoid errors")  ## enter same name of variable as in file
        t = str(input('Enter your target variable '))
        p = int(input("enter the number of predictors "))
        predict = []
        print('\n')
        for i in range(0,p):
            pred = str(input('enter the name of predictor %d: ' %(i + 1) ))

            predict.append(pred)
        pd = predict
        pd.insert(p,t)


        file2 = file[pd]
        file2 = file2.dropna()
    ##########################################################################################################################

        dataset = file2
        MSE = []

        for i in range(1 ,30):
            X = dataset.iloc[:, :-1].values ### type your predictors column here please change that -1 based on ur preference 
            y = dataset.iloc[:, -1:].values ## type your target variable index

            from sklearn.model_selection import train_test_split
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)

            from sklearn.preprocessing import StandardScaler
            scaler = StandardScaler()
            scaler.fit(X_train)

            X_train = scaler.transform(X_train)
            X_test = scaler.transform(X_test)

            from sklearn.neighbors import KNeighborsRegressor
            classifier = KNeighborsRegressor(i)
            classifier.fit(X_train, y_train)
            y_pred = classifier.predict(X_test)

            from sklearn.metrics import r2_score
            from math import sqrt
            from sklearn.metrics import mean_squared_error
            mean_squared_error(y_test ,y_pred )


            MSE.append(mean_squared_error(y_test ,y_pred )) 

            knn = MSE.index(min(MSE))+1
        print("\n")
        print("for k = " ,knn ,"algorithm showed minimum error ")




    ###########################################################################################################################        




        print(file2.head())
        from scipy.stats import zscore   ## importing library
        file_z = file2.apply(zscore)
        print('\n')
        while True:

            print('enter the predictors value of ',pd[:len(pd)-1] ,'in the same sequence')
            basket = []
            for i in range(0,len(pd)-1):
                values = float(input('enter the value of predictor %d: ' %(i +1) ))
                basket.append(values)
            b = basket


                ## converting values into z scores
            i = 0
            z = []
            for i in range(0,len(b)):
                z_val = (b[i] - file2[pd[i]].mean())/file2[pd[i]].std()
                z.append(z_val)
            z_list = z

            def euclidean(q,w):
                d = 0
                for i in range(len(q)):
                    d = d + (q[i]-w[i])**2

                return(np.sqrt(d))



            dist = []  
            for i in range(0 , len(file_z)):
                dist.append(euclidean(np.array(z_list) , file_z.iloc[i , :-1]))
            d = dist.copy()
            dist.sort()
            print('\n')
            index = []
            for i in range(0,knn):
                indx = d.index(dist[i])
                index.append(indx)
                d[indx] = -1
            print('predicted' ,pd[p] , 'is mean of Zscores of  following ', knn,' cases because this  has the lowest euclidean distance cases')
            print('\n')

            pred_zscore_mpg = np.mean(file_z.iloc[index , -1:])

                ## converting z outcome to its actual value

            mpg = round((pred_zscore_mpg*file[pd[p]].std()) + file2[pd[p]].mean() , 2) 
            print('\n')
            print('Predicted value of', pd[p] , 'is')
            print(mpg)
            
    except:  
        print("please read the statement properly and enter the appropiate details ")
```
