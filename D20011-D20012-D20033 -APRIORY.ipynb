```python
radio1 = nf.groupby(['user'])['artist']

radio1.unique()

xf = list(radio1.unique().iloc[:])
xf
```

```python
te = TransactionEncoder()
te_ary = te.fit(xf).transform(xf)
```

```python
df = pd.DataFrame(te_ary , columns = te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
f_i = apriori(df , min_support = 0.01 , use_colnames = True)
f_i
```

```python
from mlxtend.frequent_patterns import association_rules
res = association_rules(f_i , metric = "confidence" , min_threshold = 0.1)
res
```

```python
res1 = res[['antecedents' , 'consequents' ,'support', 'confidence','lift']]
res1.head(60)
```

```python
res2 = res1[res1['confidence']>=1]
res2
```
