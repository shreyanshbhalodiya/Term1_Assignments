```python
def histogram(df):
    df1 = df.select_dtypes(exclude=['object'])  
    for col in df1.columns:
        df1.hist(column = col)             # Visualizing a Histogram
        plt.xlabel(col , fontsize = 10)
        plt.ylabel("Count" , fontsize = 10)
        plt.title(col,fontsize = 15)
        #plt.show()
        plt.savefig(f"{col}.png")   # Saving the histograms in .png files in Working Directory    
```

```python
# Function to do Bi-Variate Analysis on the Variable Input By the User
def crosstab(df):
    
    # Subsetting a DataFrame on the Basis of Data type and Displaying The List of Attributes to the User.
    
        print('These are your column names with index number,select column by entering index number in box below')
        print('\n')
        basket = []  # List with All variable Index
        for i in df.columns:
            print(i , df.columns.get_loc(i))
        basket.append(df.columns.get_loc(i))
        print(basket)
    
        
     
    # Requesting the input from User for the First Argument in the Crosstab Function.
    
        x1 = " "
        masket1 = []
        while True:
            x1 = int(input('\nEnter First argument in the Crosstab function.i.e index of  variable.(Enter -1 to Close the Loop):'))
            if x1 in masket1:
                print("\nDuplicate Entry!!!! ...... Please Enter the Index Value Again ")
            masket1.append(x1)
            if x1 == -1:            # User will enter -1 as a final input
                break
            if x1<-1 or x1>basket[-1] :       #To check If user input any no. not there in Index
                print("\nOOPSSS!!!! Index out of range....... ")
            box = list(set(masket1[:]))
            new_box = []
            for i in box:
                if i>=0 and i<=basket[-1]:
                    new_box.append(i)
            int_list1 = new_box
            
        print("\nBelow are the index and variables you have selected as the First argument:", int_list1,':' ,df.columns[[int_list1]])
    
    # Requesting the input from User for the Second Argument in the Crosstab Function.
        
        x2 = " "
        masket2 = []
        while True:
            x2 = int(input('\nEnter second argument in the Crosstab function.i.e index of categorical variable.(Enter -1 to Close the Loop):'))
            if x2 in masket2:
                print("\nDuplicate Entry!!!! ...... Please Enter the Index Value Again ")
            masket2.append(x2)
            if x2 in masket1:
                print('\nSecond argument cannot have the same value as in the first argument.Please enter correct value,if you entered -1 ignore.')
            if x2 == -1:            # User will enter -1 as a final input
                break
            if x2<-1 or x2>basket[-1]:       #To check If user input any no. not there in Index
                print("\nOOPSSS!!!! Index out of range....... ")
            box = list(set(masket2[:]))
            new_box = []
            for i in box:
                if i>=0 and i<=basket[-1] and i not in masket1: #To check if user has not inputted index, which is already in argument 1 list.
                    new_box.append(i)
            int_list2 = new_box
           
            
        print("\nBelow are the index and variables you have selected as the Second argument:", int_list2,':' ,df.columns[[int_list2]])


    
        for i in df.iloc[:,int_list1]:       #Iterating First list of Variables
            for j in df.iloc[:,int_list2]:   # Iterating Second Set of Variable with Every Variable from List 1
                table = pd.crosstab(df[i],columns = df[j])
                print("\n")
                print("\n")
                rowtotal = table.sum(axis=1)    # Calculting a Row Sum
                crosstab = round(table.div(rowtotal,axis=0)*100,2)    # Calculating A Row Proportion.
                print(crosstab)
                print("\n")
                crosstab.plot(kind = 'bar' ,rot=0 , figsize = [16,5])    # Bar Plot of CrossTab
         
            
```
